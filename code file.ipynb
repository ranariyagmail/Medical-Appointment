finding Null values

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data.shape
```

```python
data = data[data['Age'] > 0]            #Remove rows with age ≤ 0
```

Remove Duplicate Rows

```python
data=data.drop_duplicates()
```

```python
data.shape
```

 Convert Date Formats to dd-mm-yyyy

```python
data['ScheduledDay'] = pd.to_datetime(data['ScheduledDay'], errors='coerce').dt.strftime('%d-%m-%Y')
data['AppointmentDay'] = pd.to_datetime(data['AppointmentDay'], errors='coerce').dt.strftime('%d-%m-%Y')
```

Rename column headers to be clean and uniform 

```python
data.columns = data.columns.str.lower().str.replace(" ", "-", regex=False)

```
